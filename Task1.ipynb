# Internship Task 1: Iris Dataset

# **Introduction and Problem Statement**

The goal of this project is to explore and visualize the Iris dataset to understand how flower measurements (sepals and petals) vary across different species.

```python
# Import the necessary toolboxes for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the built-in Iris dataset from Seaborn directly into a Pandas DataFrame
df = sns.load_dataset('iris')
```

# DATASET UNDERSTANDING

```python
# Check the dimensions of the dataset (number of rows and columns)
print("Shape:", df.shape)
```

```python
# List all the column names to understand what features we are analyzing
print("Columns:", df.columns)
```

```python
# Display the first 10 rows to visually verify the data loaded correctly
print(df.head(10))
```


****The dataset contains 150 individual flowers and 5 columns of data.****

 # **Exploratory Data Analysis (EDA) with graphs**

# Scattter Plot
**Analyzing the relationship and correlation between petal length and petal width to observe how the different flower species cluster together.**

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)

# Add a descriptive title to the graph
plt.title("Relationship:Petal Length vs Petal Width")

# Display the plot clearly without the background memory text
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)

# Add a descriptive title to the graph
plt.title("Relationship: Sepal Length vs Sepal Width")

# Display the plot clearly without the background memory text
plt.show()
```

# Histogram Plot

**Examining the frequency distribution of sepal lengths to understand the most common sizes and the overall shape of the data.**

```python
sns.histplot(df['sepal_length'], bins=15)

# Add a descriptive title to the graph
plt.title('Distribution of Sepal Lengths')

# Display the plot
plt.show()
```

```python
# df.hist() automatically makes histogram of numeric column.
df.hist(bins=15, figsize=(10, 8))

# Neat layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Box Plot

**Detecting potential outliers and comparing the median and spread of sepal widths across the three different species.**

```python
# boxplot draws a visual summary of the data's spread.
sns.boxplot(x='species', y='sepal_width', data=df)

# Add a descriptive title to the graph
plt.title('Sepal Width Ranges by Species')

# Display the plot
plt.show()
```

# Conclusion & Insights

Based on the Exploratory Data Analysis (EDA) of the Iris dataset, we can draw the following key conclusions:

1. **Feature Importance (Petals over Sepals):** Petal length and petal width are the most powerful features for distinguishing between the flower species. They show a strong positive correlation (as length increases, width increases) and create distinct clusters.
2. **The "Setosa" Isolation:** The *Setosa* species is completely separated from the other two. It is easily identifiable by its uniquely small petals and wide sepals. 
3. **Versicolor vs. Virginica:** These two species have overlapping sepal measurements, making them harder to separate using sepals alone. However, looking at their petals reveals that *Virginica* flowers are generally much larger than *Versicolor*.

**Final Takeaway:** If we were to build a machine learning model to classify these flowers, prioritizing petal measurements over sepal measurements would give us the most accurate results.
